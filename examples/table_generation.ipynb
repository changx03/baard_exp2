```python
import os
import sys

sys.path.append('..')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files_csv = np.array(os.listdir(os.path.join('..', 'csv')))
print(files_csv)
```

```python
dfs = []
files_csv = np.sort(files_csv)
print(*files_csv, sep='\n')
for f in files_csv:
    path = os.path.join('..', 'csv', f)
    df = pd.read_csv(path)
    df = df.drop(columns=['Unnamed: 0'])
    df = df.replace('baard_2stage', 'BAARD (2)')
    df = df.replace('baard_3stage', 'BAARD (3)')
    df = df.replace('lid', 'LID')
    df = df.replace('rc', 'RC')
    df = df.replace('megnet', 'MegNet')
    df = df.replace('fs', 'FS')
    dfs.append(df)
```

```python
names_pdf = [
    'banknote_nn_acc.pdf',
    'banknote_svm_acc.pdf',
    'banknote_tree_acc.pdf',
    'breastcancer_nn_acc.pdf',
    'breastcancer_svm_acc.pdf',
    'breastcancer_tree_acc.pdf',
    'cifar10_resnet_acc.pdf',
    'cifar10_vgg_acc.pdf',
    'htru2_nn_acc.pdf',
    'htru2_svm_acc.pdf',
    'htru2_tree_acc.pdf',
    'mnist_dnn_acc.pdf'
]
for i, d in enumerate(dfs):
    boxplot = d.boxplot(column=['Acc_on_adv'], by=['Defence'])
    boxplot.set_title('')
    plt.ylabel('Accuracy on Adv.')
    plt.xlabel('Defences')
    plt.suptitle('')
    plt.savefig(names_pdf[i], format='pdf', dpi=1200, bbox_inches='tight')
```
