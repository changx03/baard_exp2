```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import seaborn as sns
```

```python
cols = ["#52952c",
"#80d358",
"#328dee",
"#d7a532",
"#d5702e",
"#b166d0"]
cols_here = [cols[1], cols[5]]
```

```python
files_mnist_baard = glob.glob('result_*/mnist*baard*apgd2.*')
files_mnist_megnet = glob.glob('result_*/mnist*magnet*apgd2.*')
files_cifar_baard = glob.glob('result_*/cifar*baard*apgd.*')
files_cifar_megnet = glob.glob('result_*/cifar*magnet*apgd.*')
```

```python
dfs = []
for f in files_mnist_baard:
    df = pd.read_csv(f)
    df['Defence'] = 'BAARD'
    dfs.append(df)
for f in files_mnist_megnet:
    df = pd.read_csv(f)
    df['Defence'] = 'MagNet'
    dfs.append(df)
res_mnist = pd.concat(dfs)
res_mnist['Accuracy Difference'] = res_mnist['acc_on_adv'] - res_mnist['acc_base']
```

```python
dfs = []
for f in files_cifar_baard:
    df = pd.read_csv(f)
    df['Defence'] = 'BAARD'
    dfs.append(df)
for f in files_cifar_megnet:
    df = pd.read_csv(f)
    df['Defence'] = 'MagNet'
    dfs.append(df)
res_cifar = pd.concat(dfs)
res_cifar['Accuracy Difference'] = res_cifar['acc_on_adv'] - res_cifar['acc_base']
```

```python
# MNIST
plt.style.use('seaborn-paper')
palette = plt.get_cmap('Set2')
plt.figure(figsize=(4,3))
sns.lineplot(data=res_mnist, x="eps", y="Accuracy Difference", hue="Defence")
plt.xlabel("Epsilon")
plt.ylabel("Accuracy Improvement\n (Acc(NN + Defence) - Acc(NN))")
#plt.legend()
#plt.savefig('Plots/AdvParam_MNIST.pdf', format='pdf', dpi=1200, bbox_inches='tight')
```
